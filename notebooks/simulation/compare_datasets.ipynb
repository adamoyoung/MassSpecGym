```python
import pandas as pd
```

```python
# old = pd.read_csv("../../data/MassSpecGym_older.tsv",sep="\t")
old = pd.read_csv("../../data/MassSpecGym_old.tsv",sep="\t")
new = pd.read_csv("../../data/MassSpecGym.tsv",sep="\t")
```

```python
old
```

```python
new
```

```python
both_ids = set(old["identifier"]).intersection(set(new["identifier"]))
old_only_ids = set(old["identifier"]).difference(set(new["identifier"]))
new_only_ids = set(new["identifier"]).difference(set(old["identifier"]))
print(len(both_ids),len(new_only_ids),len(old_only_ids))
```

```python
print(old.columns.tolist())
print(new.columns.tolist())
print(set(old.columns.tolist()) - set(new.columns.tolist()))
print(set(new.columns.tolist()) - set(old.columns.tolist()))
```

```python
new_std = new.set_index("identifier")
old_std = old.set_index("identifier")
old_std = old_std[new_std.columns.values]
print((new_std.columns == old_std.columns).all())
print((new_std.index == old_std.index).all())
diff = new_std != old_std
# print(diff)
both_nan = new_std.isna() & old_std.isna()
# print(both_nan)
print((diff & ~both_nan).sum())
```

```python
print(old_std["instrument_type"].value_counts())
print(new_std["instrument_type"].value_counts())
```

```python
inst_type_diff_mask = new_std["instrument_type"] != old_std["instrument_type"]
print(old_std[inst_type_diff_mask]["instrument_type"].value_counts())
print(new_std[inst_type_diff_mask]["instrument_type"].value_counts())
```

```python
split_diff_mask = new_std["simulation_challenge"] != old_std["simulation_challenge"]
print(old_std[split_diff_mask]["simulation_challenge"])
print(new_std[split_diff_mask]["simulation_challenge"])
```

```python
old_split = old_std[old_std["simulation_challenge"]]
print(old_split.isna().sum())
print(old_split["adduct"].value_counts())
print(old_split["instrument_type"].value_counts())
```

```python
new_split = new_std[new_std["simulation_challenge"]]
print(new_split.isna().sum())
print(new_split["adduct"].value_counts())
print(new_split["instrument_type"].value_counts())
```
